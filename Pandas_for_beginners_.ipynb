```python
df['Total ATK'] = 100
```

**3.6 Condition based Updating using Apply**

```python
df['High HP'] = df['HP'].apply(lambda x: 'YES' if x > 70 else 'NO')
```

```python
df[df['High HP']=='YES'].head()
```
